```python
#Import dependencies
import pandas as pd
import json
from sqlalchemy import create_engine
import psycopg2 as ps
```

```python
# Read the csv
movie_data = pd.read_csv("./Resources/tmdb_5000_movies.csv")
movie_data.head(2)
credits_data = pd.read_csv("./Resources/tmdb_5000_credits.csv")
credits_data.head(2)
```

```python
# Clean the column genres ()
genre = []
for i in range(0,len(movie_data['genres'])):
    try:
        python_dict = json.loads(movie_data["genres"][i])
        genre.append(python_dict[0]['name'])
    except IndexError:
        genre.append(" ")

genres_df = pd.DataFrame(genre)
genres_df.head(2)
```

```python
# Clean the column production company
production_company = []
for i in range(0,len(movie_data['production_companies'])):
    try:
        python_dict = json.loads(movie_data["production_companies"][i])
        production_company.append(python_dict[0]['name'])
    except IndexError:
        production_company.append(" ")
        
production_company_df = pd.DataFrame(production_company)
production_company_df.head(2)
```

```python
# Clean the column country
country =[]
for i in range(0,len(movie_data['production_countries'])):
    try:
        python_dict = json.loads(movie_data["production_countries"][i])
        country.append(python_dict[0]['name'])
    except IndexError:
        country.append(" ")
        
country_df = pd.DataFrame(country)
country_df.head(2)
```

```python
# Clean the column cast
cast1 =[]
for i in range(0,len(credits_data['cast'])):
    try:
        python_dict = json.loads(credits_data["cast"][i])
        cast1.append(python_dict[0]['name'])
    except IndexError:
        cast1.append(" ")
        
cast1_df = pd.DataFrame(cast1)
cast1_df.head(2)
```

```python

# Clean the column cast
cast2 =[]
for i in range(0,len(credits_data['cast'])):
    try:
        python_dict = json.loads(credits_data["cast"][i])
        cast2.append(python_dict[1]['name'])
    except IndexError:
        cast2.append(" ")

cast2_df = pd.DataFrame(cast2)
cast2_df.head(2)
```

```python
# Clean the column cast
cast3 =[]
for i in range(0,len(credits_data['cast'])):
    try:
        python_dict = json.loads(credits_data["cast"][i])
        cast3.append(python_dict[2]['name'])
    except IndexError:
        cast3.append(" ")
        
cast3_df = pd.DataFrame(cast3)
cast3_df.head(2)
```

```python
# Clean the column crew and take the information that we need (Director)
director=[]
for j in range(0,len(credits_data["crew"])):
    try:
        boolean=1
        python_dict = json.loads(credits_data["crew"][j])
        for i in python_dict:
            if i["job"]=="Director" and boolean==1:
                boolean=0
                director.append(i["name"])
    except IndexError:
        director.append(" ")

director_df = pd.DataFrame(director)
director_df.head(5)
```

```python
# Clean the column crew and take the information that we need (Director of Photography)
dp=[]
for j in range(0,len(credits_data["crew"])):
    try:
        python_dict = json.loads(credits_data["crew"][j])
        for i in python_dict:
            if i["job"]=="Director of Photography":
                dp.append(i["name"])
    except IndexError:
        dp.append(" ")

dp_df = pd.DataFrame(dp)
dp_df.head(2)
```

```python
# Clean the column crew and take the information that we need (Writer)
writer=[]
for j in range(0,len(credits_data["crew"])):
    try:
        python_dict = json.loads(credits_data["crew"][j])
        for i in python_dict:
            if i["job"]=="Writer":
                writer.append(i["name"])
    except IndexError:
        writer.append(" ")

writer_df = pd.DataFrame(writer)
writer_df.head(2)
```

```python
# Clean the column crew and take the information that we need (Producer)
producer=[]
for j in range(0,len(credits_data["crew"])):
    try:
        python_dict = json.loads(credits_data["crew"][j])
        for i in python_dict:
            if i["job"]=="Producer":
                producer.append(i["name"])
    except IndexError:
        producer.append(" ")

producer_df = pd.DataFrame(producer)
producer_df.head(2)
```

```python
# Clean the column crew and take the information that we need (Music Composer)
music_composer=[]
for j in range(0,len(credits_data["crew"])):
    try:
        python_dict = json.loads(credits_data["crew"][j])
        for i in python_dict:
            if i["job"]=="Original Music Composer":
                music_composer.append(i["name"])
    except IndexError:
        music_composer.append(" ")
        
music_composer_df = pd.DataFrame(music_composer)
music_composer_df.head(2)
```

```python
# Convert both csv to DataFrames
movie_data_df = pd.DataFrame(movie_data)
credits_data_df = pd.DataFrame(credits_data)
```

```python
# fill rows that don't have information with "NaN"
movie_data_df = movie_data_df.fillna(value="NaN")
movie_data_df.count()
```

```python
movie_data_df.head(2)
```

```python
credits_data_df = credits_data_df.fillna(value="NaN")
credits_data_df.count()
```

```python
credits_data_df.head()
```

```python
# Add the columns that we need for the analisis
movie_data_df["Genre"] = genres_df
movie_data_df["Production Company"] = production_company_df
movie_data_df["Country"] = country_df
movie_data_df.head()
```

```python
# Add the columns that we need for the analisis
credits_data_df["Leading Role"] = cast1_df
credits_data_df["Supporting Role"] = cast2_df
credits_data_df["Supporting Role 2"] = cast3_df
credits_data_df["Director"] = director_df
credits_data_df["Producer"] = producer_df
credits_data_df["DP"] = dp_df
credits_data_df["Writer"] = writer_df
credits_data_df["Music Composer"] = music_composer_df
credits_data_df
```

```python
# Delete the columns that we don't need
movie_data_needed = movie_data_df.drop(['genres', 'homepage', 'keywords', 'overview',
              'popularity', 'production_companies', 'status','tagline',
              'original_title', 'production_countries', 'spoken_languages'], axis = 1)
movie_data_needed.head()
```

```python
# Delete the columns that we don't need
credits_data_needed = credits_data_df.drop(['cast', 'crew'], axis = 1)
credits_data_needed.head()
```

```python
# Rename columns
Renamed_Movie_Data_df = movie_data_needed.rename(columns = {'budget':'Budget',
                                            'original_language':'Language',
                                            'release_date':'Release_Date',
                                            'revenue':'Revenue',
                                            'runtime':'Runtime',
                                            'title':'Title',
                                            'vote_average':'Rating',
                                            'vote_count':'Vote_Count',
                                            'movie_id':'Movie_ID',
                                            'Production Company':'Production_Company',
                                            })
Renamed_Movie_Data_df.head(2)
```

```python
Renamed_Credits_Data_df = credits_data_needed.rename(columns = {
    'movie_id':'Movie_ID',
    'title':'Title',
    'Leading Role':'Leading_Role',
    'Supporting Role':'Supporting_Role',
    'Supporting Role 2':'Supporting_Role_2',
    'Music Composer':'Music_Composer'
})
Renamed_Credits_Data_df.head()
```

```python
# Reorganized the columns
organized_movie_data_df = Renamed_Movie_Data_df[["Title","Production_Company", "Genre","Country","Language","Release_Date",
                             "Runtime", "Budget", "Revenue", "Vote_Count", "Rating"]]
organized_movie_data_df.head(2)
```

```python
organized_credits_data_df = Renamed_Credits_Data_df[["Director","Leading_Role",
                             "Supporting_Role", "Supporting_Role_2", "Producer", "DP", "Writer", "Music_Composer"]]
organized_credits_data_df.head()
```
